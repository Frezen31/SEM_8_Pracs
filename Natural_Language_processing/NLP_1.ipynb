**Tasks** **on** **read** **file**

```
# storing all the words in list
with open("/content/drive/My Drive/NLP_1/it.txt", "r") as my_file_1:
  flat_list_1 = re.split(r'[\'?,-.\s:]',my_file_1.read())
while '' in flat_list_1:
  flat_list_1.remove('')

# printing the list 
print(flat_list_1)

# printing length of the list
length1 = len(flat_list_1)
print(length1)

# printing word with maximum length 
print(max(flat_list_1,key=len))
```

```

# storing all the words in list

with open("/content/drive/My Drive/NLP_1/tpv.txt", "r") as my_file:
  flat_list_2 = re.split(r'[\'?,-.\s:]',my_file.read())
while '' in flat_list_2:
  flat_list_2.remove('')

# printing the list
print(flat_list_2)

# printing length of the list
length2 = len(flat_list_2)
print(length2)

# printing word with maximum length
print(max(flat_list_2,key=len))
```

**Removing** **stopwords**

```
# defining list of stopwords
stop_words = ['a', 'an', 'the', 'of', 'is', 'this', 'that', 'to','The','A','An']
```

```
# removing stopwords for file 1
for word in flat_list_1:  
    if word in stop_words:
        flat_list_1.remove(word)

print(flat_list_1)
```

```
# removing stopwords from file 2
for word in flat_list_2: 
    if word in stop_words:
        flat_list_2.remove(word)

print(flat_list_2)
```

**Frequency** **Distribution**

```
# counts of each  word in file 1
counter1 = Counter(flat_list_1)
print(counter1)
```

```
# counts of each word in file 2
counter2 = Counter(flat_list_2)
print(counter2)
```

**Plotting** **the** **distribution**

```
# for file 1
rcParams['figure.figsize'] = [20,10]
plt.barh(*zip(*sorted(counter1.items())))
plt.show()
```

```
# for file 2
rcParams['figure.figsize'] = [20,15]
plt.barh(*zip(*sorted(counter2.items())))
plt.show()
```
